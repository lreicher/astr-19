```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)

def exp(x):
    return np.exp(x)
```

```python
y = exp(x)
plt.plot(x,y)
plt.xlabel("Time [milliseconds]")
plt.ylabel("Awesomeness")
plt.savefig("prompt1.png", bbox_inches="tight",facecolor="white")
```
