# Prompt6 - Modeling Trigonometric Functions

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# Sin(x) Function
Given a numerical input x, calculate the sin(x) and return it. X can be a single numerical value or an array of numerical values.

```python
def sin(x):
    return np.sin(x)
```

# Cos(x) Function
Given a numerical input x, calculate the cos(x) and return it. X can be a single numerical value or an array of numerical values.

```python
def cos(x):
    return np.cos(x)
```

# Tabulate the 1000 entires of x, sin(x), and cos(x)

```python
x = np.linspace(0,2*np.pi,1000)
sin_x = sin(x)
cos_x = cos(x)

print("{x} | {sin} | {cos}".format(x="x".ljust(20), sin="sin(x)".ljust(20),
                                  cos="cos(x)".ljust(20)))
print("-"*64)
for i in range(len(x)):
    print("{x} | {sin} | {cos}".format(x=str(x[i]).ljust(20),
                sin=str(sin_x[i]).ljust(20),cos=str(cos_x[i]).ljust(20)))
```

# Tabulate the first 10 entries

```python
x_arr = x[0:10]
sin_x = sin(x_arr)
cos_x = cos(x_arr)
print("{x} | {sin} | {cos}".format(x="x".ljust(20), sin="sin(x)".ljust(20),
                                  cos="cos(x)".ljust(20)))
print("-"*64)
for i in range(10):
    x_val = x_arr[i]
    print("{x} | {sin} | {cos}".format(x=str(x_val).ljust(20),
                sin=str(sin_x[i]).ljust(20),cos=str(cos_x[i]).ljust(20)))

```
